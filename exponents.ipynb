```
from IPython.display import display, Math
import sympy as sym

sym.init_printing()
```

```
x, y, z = sym.symbols('x y z')

ex = x**y * x**z
```

```
display(Math(sym.latex(ex)))
```

```
display(sym.simplify(ex))
```

```
ex1 = x**y * x**z
ex2 = x**y / x**z
ex3 = x**y * y**z

display(Math('%s \\quad %s \\quad %s' %(sym.latex(sym.simplify(ex1)), sym.latex(sym.simplify(ex2)), sym.latex(sym.simplify(ex3)))))
```

```
sym.Eq(ex1, sym.simplify(ex1))
```
