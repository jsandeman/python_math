```
import matplotlib.pyplot as plt
import math
```

```
p1 = [-4, -3]
p2 = [4, 3]

plt.plot([p1[0], p2[0]], [p1[1], p2[1]],'r')
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.show()
```

```
x = range(-20,21)

for i in x:
  plt.plot([0,i], [0, abs(i)**(1/2)])

plt.axis([-30, 30, 0, 4])
plt.show()
```

```
plt.plot([0,0], [0,2])
plt.plot([0,2], [2,2])
plt.plot([2,2], [2,0])
plt.plot([2,0], [0,0])

plt.axis('square')
plt.axis([-4, 4, -4, 4])
plt.show()
```
