```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon
```

```python
y = np.array([ [1,1], [1,4], [4,1] ])
p = Polygon(y,facecolor='m',alpha=0.5)

y1 = np.array([ [1,1], [2,3], [3,1] ])
p1 = Polygon(y1,facecolor='r',alpha=0.5)

fig,ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)
ax.set_ylim([0,4])
ax.set_xlim([0,4])

plt.show()
```
