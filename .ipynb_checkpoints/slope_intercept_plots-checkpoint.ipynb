```
import matplotlib.pyplot as plt
```

```
x = [-5, 5]
m = 0.7
b = -2

m2 = -1.25
b2 = 0.75

y = [(m*n + b) for n in x]
y2 = [(m2*n + b2) for n in x]

plt.plot(x, y)
plt.plot(x,y2)

plt.title('The Plot')

axis = plt.gca()
plt.plot([0,0], axis.get_ylim(), 'k--')
plt.plot(axis.get_xlim(), [0,0], 'k--')

plt.show()
```
