```
import numpy as np
import sympy as sym
from IPython.display import display, Math
```

```
A = np.array([ [2, 1], [3, 7] ])

print(type(A))
display(Math(sym.latex(sym.sympify(A))))
```

```
mat = np.zeros([4, 4])

for i in range(0,4):
  for j in range(0,4):
    if i == j:
      mat[i,j] = 1

display(Math(sym.latex(sym.sympify(mat))))
```

```
mul_table = np.zeros([10, 10])
dim = range(1, 11)

for a in dim:
  for b in dim:
    mul_table[a-1][b-1] = a * b

display(Math(sym.latex(sym.sympify(mul_table))))
```
