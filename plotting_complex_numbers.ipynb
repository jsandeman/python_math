```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.complex(2,3)

plt.plot(np.real(z), np.imag(z), 'ro')
plt.plot([0, np.real(z)], [0, np.imag(z)], 'r')
plt.xlabel('real')
plt.ylabel('imaginary')
plt.axis([-4,4,-4,4])
plt.grid()

plt.show()
```

```python
z1 = np.complex(-3, 1)
z2 = np.complex(-1, 1)
s = z1 + z2
p = z1 * z2
```

```python
plt.plot(np.real(z1), np.imag(z1), 'ro')
plt.plot([0, np.real(z1)], [0, np.imag(z1)], 'r')

plt.plot(np.real(z2), np.imag(z2), 'bo')
plt.plot([0, np.real(z2)], [0, np.imag(z2)], 'b')

plt.plot(np.real(s), np.imag(s), 'mo')
plt.plot([0, np.real(s)], [0, np.imag(s)], 'm')

plt.xlabel('real')
plt.ylabel('imaginary')
plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()

plt.show()
```

```python
plt.plot(np.real(z1), np.imag(z1), 'o')
plt.plot([0, np.real(z1)], [0, np.imag(z1)])

plt.plot(np.real(z2), np.imag(z2), 'o')
plt.plot([0, np.real(z2)], [0, np.imag(z2)])

plt.plot(np.real(p), np.imag(p), 'o')
plt.plot([0, np.real(p)], [0, np.imag(p)])

plt.xlabel('real')
plt.ylabel('imaginary')
plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()

plt.show()
```
