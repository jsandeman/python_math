```python
import sympy as sym
import sympy.plotting.plot as symplot
```

```python
x = sym.symbols('x')

y = x**2

p = symplot(y, show=False)
p[0].linecolor = 'm'
p.title = '$Plot \ of \ f(x) \ = x^2$'
p.show()
```

```python
x,a = sym.symbols('x,a')

expr = a/x
n = 1

p = symplot(expr.subs(a,n), show=False)
p2 = symplot(expr.subs(a,2), show=False)
p.extend(p2)

p.ylim = [-5,5]
p.xlim = [-5,5]

p[0].label = '$f(x) \ = \ %s$' %sym.latex(expr.subs(a,n))
p[0].line_color = 'r'
p[1].label = '$f(x) \ = \ %s$' %sym.latex(expr.subs(a,2))
p.title = 'Plot of a/x'
p.legend  = True
p.show()
```

```python
x,a = sym.symbols('x,a')
expr2 = a/(x**2 - a)

n = range(2,5)

p = None

colors = ['blue', 'red', 'green', 'black']

p =  symplot(expr2.subs(a,1), show=False)
p[0].line_color = colors[0]
p[0].label = '$%s$' %(sym.latex(expr2.subs(a,1)))

for i in n:
    p.extend(symplot(expr2.subs(a,i), show=False))
    p[i-1].label = '$%s$' %(sym.latex(expr2.subs(a,i)))
    p[i-1].line_color = colors[i-1]

p.legend = True
p.ylim = [-10,10]
p.xlim = [-5,5]
p.show()
```
