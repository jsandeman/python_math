```
from IPython.display import display, Math
import sympy as sym
```

```
q = sym.symbols('q')
```

```
eq1 = 3 * q + 4/q + 3 - 5 * q - 1/q - 1

simp_eq1 = sym.simplify(eq1)
display(Math(sym.latex(simp_eq1)))

sol_eq1 = sym.solve(simp_eq1, q)
display(Math(sym.latex(sol_eq1)))
```

```
eq2 = 2 * q + 3 * q**2 - 5/q - 4/q**3

simp_eq2 = sym.simplify(eq2)
can_eq2 = sym.cancel(simp_eq2)

display(Math(sym.latex(can_eq2)))
```

```
eq3 = (3**(1/2) + 15**(1/2) * q) / (2**(1/2) + 10**(1/2) * q)

simp_eq3 = sym.simplify(eq3)

display(Math(sym.latex(simp_eq3)))

simp_eq3.subs(q, 4)
```
