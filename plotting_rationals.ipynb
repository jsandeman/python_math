```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3,3,100)
y = 2 - x**2

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-4,4,200)
e = range(-1, 4)

for i in e:
    y = x**i
    plt.plot(x,y,label='$y = x^{%s}$' %i,linewidth=2)

plt.legend()
plt.xlim([x[0], x[-1]])
plt.ylim([-20,20])
plt.show()
```
