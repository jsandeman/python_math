```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon
```

```python
y = np.array([ [1,1], [1,4], [4,1] ])
p = Polygon(y,facecolor='m',alpha=0.5)

y1 = np.array([ [1,1], [2,3], [3,1] ])
p1 = Polygon(y1,facecolor='r',alpha=0.5)

fig,ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)
ax.set_ylim([0,4])
ax.set_xlim([0,4])

plt.show()
```

```python
f = [[x,-(x**2)] for x in np.linspace(-2.0, 2.0,100)]
s = [ [-0.5,-4.0], [-0.5,-2.5], [0.5,-2.5], [0.5,-4.0]]

parab = Polygon(f,facecolor='g',alpha=0.3,edgecolor='k')
square = Polygon(s,facecolor='k')

fig,ax = plt.subplots()
ax.add_patch(parab)
ax.add_patch(square)
ax.set_ylim([-4.0,0.0])
ax.set_xlim([-2.0,2.0])
plt.axis('off')

plt.show()
```
