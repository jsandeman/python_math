```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [ [1,2], [1,4]]

plt.imshow(A)
plt.xticks([0,1])
plt.yticks([0,1])

plt.show()
```

```python
B = [[3*i - 4*j for j in range(14)] for i in range(10)]

plt.imshow(B)

for i in range(10):
    for j in range(14):
        plt.text(j, i, int(B[i][j]), horizontalalignment='center', verticalalignment='center')
        
plt.set_cmap('plasma')

plt.show()
```

```python
matrix_size = 10

C = [ [(-1)**(i + j) for j in range(matrix_size)] for i in range(matrix_size)]

fig = plt.imshow(C)
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
plt.set_cmap('binary')

plt.show()
```
