```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3,3,100)

plt.plot(x, np.exp(x))
plt.xlabel('x')
plt.ylabel('$e^x$')
plt.show()

plt.plot(x, np.log(x))
plt.xlabel('x')
plt.ylabel('log(x)')
plt.show()
```

```python
x = np.linspace(0.1,10)
y1 = np.exp(x)
y2 = np.log(np.exp(y1))

plt.plot(y1,y2,'-s')
plt.axis('square')
plt.xlabel('$e^x$')
plt.ylabel('$e^\\log(x)$')
plt.axis([0,10,0,10])

plt.show()
```
