```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = [23, 45,32, 2, 16]

s = np.sum(a)

p = np.prod(a)

c = np.cumsum(a)

print(s)
print(p)
print(c)

plt.plot(a, 'bo-',label='data set')
plt.plot(c, 'rs-',label='cumulative sums')
plt.legend()
plt.show()
```

```python
a = [1, 2, 3, 4, 5]

a2 = [i**2 for i in a]

print(sum(a)/sum(a2) == 1/sum(a))

print(np.prod(a)/np.prod(a2) == 1/np.prod(a))
```

```python
d = [6, 11, 18, 32]
diff = np.diff(d)
diff2 = np.diff(d,2)
print(diff)
print(diff2)
```

```python
x = np.linspace(-2,2,50)
x2 = x**2
dx = np.diff(x2) / np.diff(x)

plt.plot(x, x2, 'r', label='f(x) = x^2')
plt.plot(x[1:], dx, 'b', label='df/dx')
#plt.xlim([-3.0,3.0])
plt.ylim([-1.0, 2.0])
plt.legend()
plt.show()
```
